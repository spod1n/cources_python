##### Matplotlib - бібліотека для візуалізації. 
Має можливість створювати: статичні, анімовані, інтерактивні графіки

У Matplotlib доступно декілька інтерфейсів для створення графіків:
> MATLAB-style (pyplot API), який є найпростішим у використанні та найбільш поширеним \

> **linewidth** - товщина лінії, linewidth  = 1.5  
> **color**   - колір лінії, приймає як колір так і інші формати(шістнадцятковий код, кортежі color)  
> **linestyle** - стиль лінії : '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted    
> **alpha** - прозорість   
> **marker** -  символ  "o" - коло, "s" - квадрат, "^" - трикутник, "x"

`````python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

iris = sns.load_dataset('iris')
iris.head()
`````

`````python
X = iris.drop('species', axis = 1)
y = iris['species']
`````

`````python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42, test_size= 0.2)
`````

`````python
model = KNeighborsClassifier(n_neighbors= 7)
model.fit(X_train, y_train)
`````

`````python
y_pred = model.predict(X_test)
`````

`````python
print(accuracy_score(y_test, y_pred))
`````
